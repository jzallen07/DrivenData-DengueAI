### Examine the missing data

```python
# missing data for the training data
df_train.isnull().sum()
```

```python
# missing values for the testing features
df_test = pd.read_csv('dengue_features_test.csv')
df_test.isnull().sum()
```

So, know that we have established that there are missing data, how should we progress? This seems to be one of the perienial questions of data science. Missing data in real-world scenarios in simply a fact of life (and representative of it). In this case I will be imputing the missing variables as the do not reperesent a significant proportion of the data fields and any real-world delployable pipeline would beed the ability to cope with missing data points. Further, as we are dealing with semi-continuous field like rain fall and temperature this is not an uncommon or inaccuate practice. Imputing data is obviously not a perfect solution and does hender the absolute (theoretical output of the model) but in this case I think the belefits outweight the risks. Plus, if the model is not performing well durring validation we can always return to this point and use another or modified solution.

Given that we are working with two distinct geographic locations it will be necessary to split the feature data for imputing.

```python
df_sj = df_train[df_train['city'] == 'sj']
print('sj', df_sj.shape)
df_iq = df_train[df_train['city'] == 'iq']
print('iq', df_iq.shape)
```

I would like to thank ### for the inventive way of displaying the holes in the data.

```python
# viz missing data

fig, ax = plt.subplots(figsize=(8,12))
sns.heatmap(df_iq.isnull().reset_index(drop=True),ax=ax, cbar = False, yticklabels = 50)

plt.ylabel("Row", size = 22)
plt.xlabel("Feature", size = 22)
plt.title("Iquitos Missing Data", size = 32)
```

```python
# viz missing data

fig, ax = plt.subplots(figsize=(8,12))
sns.heatmap(df_sj.isnull(),ax=ax, cbar = False, yticklabels = 50)

plt.ylabel("Row", size = 22)
plt.xlabel("Feature", size = 22)
plt.title("San Juan Missing Data", size = 32)
```

Given the semi-continuous nature of many of these features and the seasonal shift in 
many of them e.g. rainfall it makes more sense to use a forward or backfill approach from the nearest value rather than a mean or median approach. This is unless we wish to get into a much more intensive imputing process which would account for seasonal i.e. monthly averages. For now, I will save that for another time.

```python
df_sj = df_sj.fillna(method = 'ffill')
df_iq = df_iq.fillna(method = 'ffill')
print('San Juan:',df_sj.isnull().any().sum())
print('Iquitos:',df_iq.isnull().any().sum())
```

Now that our feature set is mostly cleaned-up, lets take a closer look at the target data.

```python
# let take anothe look at the weekly fluctuation in each city
plt.figure(figsize=(12,6))
sns.lineplot(x='weekofyear', y='total_cases', data=target, hue='city')
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(df_iq.corr()**2)
```

```python
df_sj.set_index('week_start_date', drop = True, inplace = True)
df_iq.set_index('week_start_date', drop = True, inplace = True)
df_sj['nvdi_mean'] = df_sj[df_sj.columns[3:7]].mean(axis = 1)
df_iq['nvdi_mean'] = df_iq[df_iq.columns[3:7]].mean(axis = 1)
```

```python
df_sj.info()
```

```python
# plot week by week and yearly average for San Juan
fig, ax = plt.subplots(2, 1, figsize=(22, 12))

ndvi_cols = ['nvdi_mean', 'ndvi_ne', 'ndvi_nw', 'ndvi_se', 'ndvi_sw']

#top plot
plt.subplot(211)
for i in ndvi_cols:
    sns.lineplot(y=df_sj[i], x=df_sj.index, data=df_sj, markers=True)
plt.title('NDVI Levels San Juan, Puerto Rico', size = 25)
plt.xlabel('Year', size = 20)
plt.ylabel('NDVI', size = 20)

# bottom plot
plt.subplot(212)
for i in ndvi_cols:
    sns.lineplot(y=df_iq[i], x=df_iq.index, data=df_iq, markers=True)
plt.title('NDVI Levels in Iquitos, Peru', size = 25)
plt.xlabel('Week of Year', size = 20)
plt.ylabel('NDVI', size = 20)
plt.legend(loc = 'best')
    
plt.tight_layout(pad=3)
```

```python
# re-load and re-process the data to ensure a clean output
df = pd.read_csv('dengue_features_train.csv', parse_dates=[3])
target = pd.read_csv('dengue_labels_train.csv')

# merge on the target variable
df['total_cases'] = target['total_cases']

# split by city
df_sj = df[df['city'] == 'sj']
df_iq = df[df['city'] == 'iq']

## forward fill the mising values
df_sj = df_sj.fillna(method = 'ffill')
df_iq = df_iq.fillna(method = 'ffill')

#drop non-numerical values
df_sj.drop(['city', 'year'], axis = 1, inplace = True)
df_iq.drop(['city', 'year'], axis = 1, inplace = True)

# export the cleaned data to csv
df_iq.to_csv(path_or_buf='df_iq.csv', index=False)
df_sj.to_csv(path_or_buf='df_sj.csv', index=False)
```
